### There is no Block Scope in python:
`
Inside conditional statements if we define a variable it has a global scope
`

````python
game_level = 3
enemies = ["skeleton", "zombies", "aliens"]
if game_level < 5:
    new_enemy = enemies[0]
print(new_enemy)
````

````python
game_level = 3
def create_enemy():
    enemies = ["skeleton", "zombies", "aliens"]
    if game_level < 5:
        enemy_2 = enemies[0]
    print(enemy_2)
create_enemy()
````

````python
enemies = "outside_func"

def increase_enemies():
  enemies = "inside_func"
  print(f"enemies inside function: {enemies}")

increase_enemies()
print(f"enemies outside function: {enemies}")
````

````python
enemies = 1

def increase_enemies(): # enemies are not accesed by function
                        # either pass as argument or define it inside a func
  enemies = 1
  enemies +=1
  print(f"enemies inside function: {enemies}")

increase_enemies()
print(f"enemies outside function: {enemies}")
````

````python
enemies = 1

def increase_enemies(): # enemies are not accesed by function
                        # either pass as argument or define it inside a func
  global enemies
  enemies +=1
  print(f"enemies inside function: {enemies}")

increase_enemies()
print(f"enemies outside function: {enemies}")
````

````python
enemies = 10

def increase_enemies():
  
  print(f"enemies inside function: {enemies}")
  return enemies + 1

print(increase_enemies())
print(f"enemies outside function: {enemies}")

````
